```python
import pandas as pd

# Fake dataset consist of few sentences.
data = ['Hello, how are you!',
        'Win money, win from home.',
        'Call me now.',
        'My heart will go on and on....',
        'France Vs Croatia - FIFA Worldcup 2018 Final!',
        'Hello, Call hello you tomorrow?',
        'Hey, didnt I gave you his number']
```

```python
# Data preprocessing with CountVectorizer()
# Import the CountVectorizer method and create an instance of it called 'count_vector'. 

from sklearn.feature_extraction.text import CountVectorizer
count_vector = CountVectorizer()
```

```python
# fit the data to CountVectorizer object you have created using fit()
count_vector.fit(data)
count_vector.get_feature_names()
```

```python
# Create data tokenised array
doc_array = count_vector.transform(data).toarray()
doc_array
```

```python
# Create a panda dataframe for doc_array
frequency_matrix = pd.DataFrame(doc_array, columns = count_vector.get_feature_names())
frequency_matrix
```
